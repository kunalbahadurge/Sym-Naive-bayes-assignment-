# **Theoretical Questions**

    Q1. What is a Support Vector Machine (SVM)
    = A Support Vector Machine (SVM) is a supervised machine learning algorithm used mainly for classification and sometimes regression tasks, which works by finding the best boundary, called a hyperplane, that separates data points of different classes in a dataset.
    For example, in a medical diagnosis system, an SVM model can classify patients as diseased or healthy based on features such as age, blood pressure, and test results, helping doctors make faster and more accurate decisions.


    Q2. What is the difference between Hard Margin and Soft Margin SVM
    = The difference between Hard Margin and Soft Margin SVM lies in how strictly the model separates the data. Hard Margin SVM tries to find a decision boundary that perfectly separates the classes with no misclassification, meaning all data points must lie on the correct side of the margin, which works well only when the data is completely linearly separable and free from noise. However, in most real-world datasets perfect separation is rare, so Soft Margin SVM is used, which allows some data points to be misclassified or fall inside the margin in order to create a more flexible and generalized boundary. This helps the model handle noisy or overlapping data better. For example, in a medical dataset predicting whether a patient has a disease, some patient records may overlap in features, so a soft margin SVM can still create a useful decision boundary even when perfect separation is not possible.


    Q3.  What is the mathematical intuition behind SVM
    = The mathematical intuition behind a **Support Vector Machine (SVM)** is to find the best decision boundary (called a **hyperplane**) that separates two classes of data while keeping the **maximum possible margin** between the boundary and the nearest data points from each class. These nearest points are called **support vectors**, and they determine the position of the boundary. By maximizing this margin, SVM aims to create a model that generalizes well and makes accurate predictions on new data.

    y = w^T x + b
    In this equation, (w) represents the weight vector, (x) represents the input features, and (b) is the bias term that shifts the boundary. For example, in an email classification system, SVM can separate spam and non-spam emails based on features like word frequency, creating a clear boundary that helps in accurate classification.


    Q4. What is the role of Lagrange Multipliers in SVM
    = The role of Lagrange multipliers in a Support Vector Machine (SVM) is to help solve the optimization problem of finding the hyperplane that maximizes the margin between classes while satisfying certain constraints. Instead of directly solving the constrained optimization problem, SVM uses Lagrange multipliers to convert it into a form that can be solved more efficiently. During this process, each training point gets an associated multiplier value, and only the points with non-zero multipliers become support vectors, meaning they are the critical points that determine the final decision boundary.


    Q5. What are Support Vectors in SVM
    = Support Vectors are the data points in a Support Vector Machine (SVM) that lie closest to the decision boundary (hyperplane) and directly determine the position and orientation of that boundary. These points are the most critical elements of the dataset because the SVM algorithm uses them to maximize the margin between different classes, while other data points that are farther away do not significantly influence the model. In simple terms, support vectors are the key observations that “support” the creation of the separating boundary. For example, in a medical dataset used to classify patients as diseased or healthy, the patients whose feature values lie closest to the dividing line between the two groups will become the support vectors and will strongly influence how the model separates the classes.

    
    Q6. What is a Support Vector Classifier (SVC)?
    = A Support Vector Classifier (SVC) is a supervised machine learning algorithm based on Support Vector Machines that is used for classification tasks, where the goal is to separate data points into different classes using an optimal decision boundary called a hyperplane. The model finds the boundary that maximizes the margin between classes, and the data points closest to the boundary are called support vectors, which determine the final position of the boundary. SVC can also use kernel functions to handle non-linear data by transforming it into higher-dimensional space where it becomes separable. For example, an SVC model can classify emails as spam or not spam based on features like word frequency and message length.

    Q7. What is a Support Vector Regressor (SVR)?
    = A Support Vector Regressor (SVR) is a variation of the Support Vector Machine used for regression problems, where the goal is to predict continuous numerical values instead of categories. Instead of finding a boundary between classes, SVR tries to fit a function that keeps prediction errors within a certain acceptable margin while maintaining the largest possible margin from the data points. It focuses only on important points near the boundary, similar to support vectors in classification. For example, SVR can be used to predict house prices based on features such as size, location, and number of rooms.


    Q8.  What is the Kernel Trick in SVM?
    =The Kernel Trick in SVM is a technique used to handle non-linearly separable data by transforming the original data into a higher-dimensional space where a linear boundary can separate the classes. Instead of explicitly performing this transformation, the kernel function calculates the similarity between data points directly in the higher dimension, making the computation more efficient. This allows SVM to solve complex classification problems without heavy calculations. For example, in image classification where classes overlap in the original feature space, the kernel trick helps the model find a separating boundary in a higher-dimensional space.

    Q9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel
    = The Linear Kernel is used when the data is linearly separable and works by finding a straight-line decision boundary, making it simple and computationally efficient. The Polynomial Kernel creates curved boundaries by considering polynomial relationships between features, which helps model more complex patterns. The RBF (Radial Basis Function) Kernel is the most flexible and widely used kernel because it can model highly non-linear relationships by creating complex decision boundaries around data points. For example, text classification may work well with a linear kernel, while image recognition problems often benefit from the RBF kernel.

    Q10. What is the effect of the C parameter in SVM?
    = The C parameter in SVM controls the trade-off between maximizing the margin and minimizing classification errors. A large C value tries to classify all training points correctly, which can lead to overfitting because the model becomes sensitive to noise. A small C value allows some misclassifications but focuses on creating a wider margin, which can improve generalization. For example, in a medical diagnosis dataset with noisy measurements, using a smaller C can help the model avoid overfitting to unusual patient records.

    Q11. What is the role of the Gamma parameter in RBF Kernel SVM?
    = The Gamma parameter in an RBF Kernel SVM determines how far the influence of a single training data point reaches when forming the decision boundary. A high gamma value makes the model focus only on very close data points, leading to complex boundaries and possible overfitting. A low gamma value considers points farther away, resulting in smoother and simpler boundaries. For example, in image classification, a moderate gamma helps capture patterns without fitting noise.


    Q12. What is the Naive Bayes classifier, and why is it called "Naive"?
    = The Naive Bayes classifier is a probabilistic machine learning algorithm based on Bayes’ Theorem that is used for classification tasks by calculating the probability that a data point belongs to a particular class. It assumes that all input features are independent of each other, which is a strong and often unrealistic assumption, and this is why it is called “Naive.” Despite this simplification, the algorithm performs well in many real-world applications. For example, Naive Bayes is widely used in spam email detection, where the presence of certain words helps predict whether an email is spam.


    Q13. What is Bayes’ Theorem?
    = Bayes’ Theorem is a mathematical formula used to calculate the probability of an event based on prior knowledge of related conditions. It describes how the probability of a hypothesis changes when new evidence is observed, combining prior probability with likelihood to compute posterior probability. In machine learning, it helps determine the probability that a data point belongs to a specific class. For example, in medical diagnosis, Bayes’ theorem can help estimate the probability that a patient has a disease based on symptoms and previous medical data.


    Q14. Differences between Gaussian, Multinomial, and Bernoulli Naive Bayes
    = Gaussian Naive Bayes is used when features are continuous and assumed to follow a normal distribution, such as measurements like height or temperature. Multinomial Naive Bayes is mainly used for discrete counts like word frequencies in text classification problems. Bernoulli Naive Bayes works with binary features that represent presence or absence of a feature. For example, Gaussian NB may predict medical test results, Multinomial NB is commonly used in spam detection, and Bernoulli NB checks whether certain words appear in a message.


    Q15. When should you use Gaussian Naive Bayes?
    =Gaussian Naive Bayes should be used when the input features are continuous numerical values and can reasonably be assumed to follow a normal distribution. It works well when the dataset contains measurements such as age, weight, blood pressure, or sensor readings. Because it calculates probabilities based on mean and variance, it is particularly useful in scientific and medical datasets. For example, it can classify patients as healthy or sick based on continuous medical measurements.


    Q16. Key assumptions made by Naive Bayes
    = The main assumption made by the Naive Bayes classifier is feature independence, meaning each feature contributes to the prediction independently of the others given the class label. It also assumes that all features have equal importance and that the probability distributions used for modeling the data are correct. While these assumptions are often unrealistic, the model still performs surprisingly well in many practical situations. For example, in spam detection, the algorithm assumes that the occurrence of each word is independent of others.


    Q17. Advantages and disadvantages of Naive Bayes
    = Naive Bayes has several advantages such as fast training speed, low computational cost, and good performance on large datasets, especially in text classification problems. It also works well even with limited training data. However, its main disadvantage is the unrealistic assumption that features are independent, which may reduce accuracy in complex datasets. For example, in sentiment analysis, words in a sentence influence each other, but Naive Bayes still treats them as independent.


    Q18. Why is Naive Bayes good for text classification?
    = Naive Bayes is particularly effective for text classification because text data often contains a large number of features (words), and the algorithm handles high-dimensional data efficiently. It calculates probabilities of word occurrences for each class and can quickly determine which class a document most likely belongs to. It also performs well even with relatively small training datasets. For example, Naive Bayes is commonly used for spam filtering, sentiment analysis, and news categorization.


    Q19. Compare SVM and Naive Bayes for classification tasks
    = SVM and Naive Bayes are both used for classification but differ in how they learn patterns from data. SVM is a margin-based algorithm that finds an optimal decision boundary between classes and often performs well in high-dimensional datasets but may require more computational resources. Naive Bayes is a probabilistic algorithm that uses probability calculations based on Bayes’ theorem and is faster and simpler but relies on strong independence assumptions. For example, Naive Bayes is widely used in text classification, while SVM is often used in image recognition or complex pattern classification.



    Q20. How does Laplace Smoothing help in Naive Bayes?
    = Laplace Smoothing is used in Naive Bayes to handle the problem of zero probability, which occurs when a feature value does not appear in the training dataset for a particular class. Without smoothing, the entire probability calculation could become zero, leading to incorrect predictions. Laplace smoothing solves this by adding a small constant value (usually 1) to all feature counts, ensuring that every possible feature has a non-zero probability. For example, in spam classification, if a new email contains a word that never appeared in spam training data, Laplace smoothing prevents the probability of spam from becoming zero.

Q24. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision
boundary

```python
from sklearn.datasets import make_blobs
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np

X,y = make_blobs(n_samples=200,centers=2,random_state=42)

model = SVC(kernel="poly",degree=3)
model.fit(X,y)

plt.scatter(X[:,0],X[:,1],c=y)

ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx = np.linspace(xlim[0],xlim[1],30)
yy = np.linspace(ylim[0],ylim[1],30)

YY,XX = np.meshgrid(yy,xx)
xy = np.vstack([XX.ravel(),YY.ravel()]).T

Z = model.decision_function(xy).reshape(XX.shape)

ax.contour(XX,YY,Z,levels=[0])

plt.show()
```

Q27. Write a Python program to train an SVM Classifier with different C values and compare the decision
boundaries visually

```python
from sklearn.datasets import make_blobs
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np

X,y = make_blobs(n_samples=200,centers=2,random_state=42)

C_values = [0.1,1,10]

for C in C_values:
    model = SVC(kernel="linear",C=C)
    model.fit(X,y)

    plt.figure()

    plt.scatter(X[:,0],X[:,1],c=y)

    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    xx = np.linspace(xlim[0],xlim[1],30)
    yy = np.linspace(ylim[0],ylim[1],30)

    YY,XX = np.meshgrid(yy,xx)
    xy = np.vstack([XX.ravel(),YY.ravel()]).T

    Z = model.decision_function(xy).reshape(XX.shape)

    ax.contour(XX,YY,Z,levels=[0])

    plt.title("C = " + str(C))
    plt.show()
```

Q31. Write a Python program to train an SVM Classifier and use GridSearchCV to tune the hyperparameters (C,
gamma, kernel)

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

data = load_breast_cancer()
X = data.data
y = data.target

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)

param_grid = {
    "C":[0.1,1,10],
    "gamma":[0.001,0.01,0.1],
    "kernel":["linear","rbf"]
}

grid = GridSearchCV(SVC(),param_grid,cv=5)
grid.fit(X_train,y_train)

best_model = grid.best_estimator_

pred = best_model.predict(X_test)

print("Best Parameters:",grid.best_params_)
print("Accuracy:",accuracy_score(y_test,pred))
```

Q40.  Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and
compare accuracy

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score

data = load_breast_cancer()
X = data.data
y = data.target

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)

model1 = SVC(kernel="linear")
model1.fit(X_train,y_train)
pred1 = model1.predict(X_test)
print("Accuracy without RFE:",accuracy_score(y_test,pred1))

selector = RFE(SVC(kernel="linear"),n_features_to_select=10)
X_train_sel = selector.fit_transform(X_train,y_train)
X_test_sel = selector.transform(X_test)

model2 = SVC(kernel="linear")
model2.fit(X_train_sel,y_train)
pred2 = model2.predict(X_test_sel)
print("Accuracy with RFE:",accuracy_score(y_test,pred2))
```

Q46. Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

data = load_breast_cancer()
X = data.data
y = data.target

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)

model = SVC(probability=True)
model.fit(X_train,y_train)

prob = model.predict_proba(X_test)[:,1]

precision,recall,_ = precision_recall_curve(y_test,prob)

plt.plot(recall,precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.show()
```
